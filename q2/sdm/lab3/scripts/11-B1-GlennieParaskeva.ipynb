## Imports

```python
import pandas as pd

from rdflib import Graph, Namespace
from rdflib.namespace import RDFS, XSD, RDF, OWL
```

## Creating Graph

```python
onto = Namespace('http://example.com/sdm_lab3#')

g = Graph()

# Paper subclasses
g.add((onto.FullPaper, RDFS.subClassOf, onto.Paper))
g.add((onto.ShortPaper, RDFS.subClassOf, onto.Paper))
g.add((onto.DemoPaper, RDFS.subClassOf, onto.Paper))
g.add((onto.Poster, RDFS.subClassOf, onto.Paper))

# Venue subclasses
g.add((onto.Conference, RDFS.subClassOf, onto.Venue))
g.add((onto.Journal, RDFS.subClassOf, onto.Venue))

# Conferences subclasses
g.add((onto.Workshop, RDFS.subClassOf, onto.Conference))
g.add((onto.ExpertGroup, RDFS.subClassOf, onto.Conference))
g.add((onto.RegularConference, RDFS.subClassOf, onto.Conference))
g.add((onto.Symposium, RDFS.subClassOf, onto.Conference))

# Person subclasses
g.add((onto.Author, RDFS.subClassOf, onto.Person))
g.add((onto.Reviewer, RDFS.subClassOf, onto.Person))
g.add((onto.Responsible, RDFS.subClassOf, onto.Person))

# Responsible subclasses
g.add((onto.Editor, RDFS.subClassOf, onto.Responsible))
g.add((onto.Chairperson, RDFS.subClassOf, onto.Responsible))

# Attributes
g.add((onto.decision, RDFS.domain, onto.Review))
g.add((onto.decision, RDFS.range, XSD.boolean))

g.add((onto.reviewText, RDFS.domain, onto.Review))
g.add((onto.reviewText, RDFS.range, XSD.string))

g.add((onto.personName, RDFS.domain, onto.Person))
g.add((onto.personName, RDFS.range, XSD.string))

g.add((onto.title, RDFS.domain, onto.Paper))
g.add((onto.title, RDFS.range, XSD.string))

g.add((onto.keywords, RDFS.domain, onto.Area))
g.add((onto.keywords, RDFS.range, XSD.string))

g.add((onto.proceedingsYear, RDFS.domain, onto.Proceedings))
g.add((onto.proceedingsYear, RDFS.range, XSD.int))

g.add((onto.volumeYear, RDFS.domain, onto.Volume))
g.add((onto.volumeYear, RDFS.range, XSD.int))

g.add((onto.venueName, RDFS.domain, onto.Venue))
g.add((onto.venueName, RDFS.range, XSD.string))

# Relationships
g.add((onto.submits, RDFS.domain, onto.Author))
g.add((onto.submits, RDFS.range ,onto.Paper))

g.add((onto.handsIn, RDFS.domain, onto.Reviewer))
g.add((onto.handsIn, RDFS.range, onto.Review))

g.add((onto.assignedTo, RDFS.domain, onto.Reviewer))
g.add((onto.assignedTo, RDFS.range, onto.Paper))

g.add((onto.assigns, RDFS.domain, onto.Responsible))
g.add((onto.assigns, RDFS.range, onto.Reviewer))

g.add((onto.includedInVolume, RDFS.domain, onto.Paper))
g.add((onto.includedInVolume, RDFS.range, onto.Volume))

g.add((onto.includedInProceedings, RDFS.domain, onto.Paper))
g.add((onto.includedInProceedings, RDFS.range, onto.Proceedings))

g.add((onto.submittedTo, RDFS.domain, onto.Paper))
g.add((onto.submittedTo, RDFS.range, onto.Venue))

g.add((onto.paperRelatedTo, RDFS.domain, onto.Paper))
g.add((onto.paperRelatedTo, RDFS.range, onto.Area))

g.add((onto.venueRelatedTo, RDFS.domain, onto.Venue))
g.add((onto.venueRelatedTo, RDFS.range, onto.Area))

g.add((onto.isPartOfJournal, RDFS.domain, onto.Volume))
g.add((onto.isPartOfJournal, RDFS.range, onto.Journal))

g.add((onto.isPartOfConference, RDFS.domain, onto.Proceedings))
g.add((onto.isPartOfConference, RDFS.range, onto.Conference))

g.add((onto.handledByChairperson, RDFS.domain, onto.Conference))
g.add((onto.handledByChairperson, RDFS.range, onto.Chairperson))

g.add((onto.handledByEditor, RDFS.domain, onto.Journal))
g.add((onto.handledByEditor, RDFS.range, onto.Editor))

g.serialize('11-B1-GlennieParaskeva.rdf',format='xml')
```
